```python
import pandas as pd
file1 = pd.read_csv("t2_registry 20190619.csv")
file2 = pd.read_csv("t2_ec 20190619.csv")
```

```python
file1.head()
```

```python
file2.head()
```

```python
merged = pd.merge(left=file1, right=file2, on=['RID','VISCODE'], how='left')
merged.head()
```

```python
columns = merged[['ID_x', 'RID', 'USERID_x', 'VISCODE', 'SVDOSE', 'ECSDSTXT']]
columns.columns = ['ID', 'RID', 'USERID', 'VISCODE', 'SVDOSE', 'ECSDSTXT']
columns.head()
```

```python
filtered = columns[(columns['VISCODE']  == 'w02') & (columns['SVDOSE'] == 'Y') & (columns['ECSDSTXT'] != 280)]
filtered.head()
```

```python
filtered.to_csv('results.csv', columns=['RID', 'USERID', 'VISCODE', 'SVDOSE', 'ECSDSTXT'], index=False)
```
